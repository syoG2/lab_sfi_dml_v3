```python
import sys
from pathlib import Path
```

```python
# aggregate_scoreで実行した結果をmarkdownの表形式で出力する

import pandas as pd
import json

key = [
    "ave-alpha",
    "ave-n_pred_lus",
    "ave-n_pred_clusters",
    "ave-pu",
    "ave-ipu",
    "ave-puf",
    "ave-bcp",
    "ave-bcr",
    "ave-bcf",
]

model_names = ["vanilla","siamese_distance","triplet_distance","softmax_classification","arcface_classification","adacos_classification"]
clusterings = ["onestep-average","twostep-xmeans-average"]

print("|" + "|".join(["model_name"] + key) + "|")
divider = ["----" for k in key ]
divider.insert(0, "----")
divider = "|" + "|".join(divider) + "|"
print(divider)
for clustering in clusterings:
    for model_name in model_names:

        input_file = Path(f"../data/verb_clustering/aggregate_scores_clustering/bert-base-uncased/{model_name}/wm/{clustering}/metrics_test.json")

        if not input_file.exists():
            row = ["-" for k in key ]
            row.insert(0, model_name)
            row = "|" + "|".join(row) + "|"
            print(row)
            continue
        with open(input_file, 'r') as f:
            data = json.load(f)

        row = [str(data[k]) if k in data else "-" for k in key ]
        if clustering == "onestep-average":
            row.insert(0, model_name + "_1")
        else:
            row.insert(0, model_name + "_2")
        row = "|" + "|".join(row) + "|"
        print(row)
print()

add_methods = ["ratio","sequential"]
c4_rates = [0,1,2]
# c4_rates = [0]
for add_method in add_methods:
    for c4_rate in c4_rates:
        print(f"add_method:{add_method}")
        print(f"c4_rate:{c4_rate}")
        print("|" + "|".join(["model_name"] + key) + "|")
        divider = ["----" for k in key ]
        divider.insert(0, "----")
        divider = "|" + "|".join(divider) + "|"
        print(divider)
        for clustering in clusterings:
            for model_name in model_names:

                input_file = Path(f"../data/verb_clustering_c4/aggregate_scores_clustering/{add_method}/{c4_rate}/bert-base-uncased/{model_name}/wm/{clustering}/metrics_test.json")

                if not input_file.exists():
                    row = ["-" for k in key ]
                    row.insert(0, model_name)
                    row = "|" + "|".join(row) + "|"
                    print(row)
                    continue
                with open(input_file, 'r') as f:
                    data = json.load(f)

                row = [str(data[k]) if k in data else "-" for k in key ]
                if clustering == "onestep-average":
                    row.insert(0, model_name + "_1")
                else:
                    row.insert(0, model_name + "_2")
                row = "|" + "|".join(row) + "|"
                print(row)
        print()
    print()
```
