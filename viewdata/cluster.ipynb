```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
# ほとんどがC4の用例からなるクラスタを調べる

pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)

settings = ["all_3_0","all_3_1","all_3_2"]
add_methods = ["ratio","sequential"]
c4_rates = [0,1,2]
input_file = Path(f"../data/verb_clustering_c4/clustering/{add_methods[0]}/{c4_rates[1]}/{settings[0]}/bert-base-uncased/adacos_classification/wm/twostep-xmeans-average/exemplars_test.jsonl")
df = pd.read_json(input_file, lines=True)


# frame_clusterごとにsourceがc4の割合とサイズを計算
c4_ratio = df.groupby('frame_cluster')['source'].apply(lambda x: (x == 'c4').mean())
cluster_size = df.groupby('frame_cluster').size()
framenet_size = df.groupby('frame_cluster')['source'].apply(lambda x: (x == 'framenet').sum())
c4_size = df.groupby('frame_cluster')['source'].apply(lambda x: (x == 'c4').sum())

# データフレームにまとめる
c4_ratio_df = pd.DataFrame({
    'c4_ratio': c4_ratio,
    'size': cluster_size,
    'framenet_size': framenet_size,
    'c4_size': c4_size
})
# c4_ratioで降順ソート、同率の場合はsizeで降順ソート
c4_ratio_sorted = c4_ratio_df.sort_values(by=['c4_ratio', 'size'], ascending=[False, False])

# 結果を表示
print(c4_ratio_sorted)
```
