```python
from nltk.corpus import framenet
import pandas as pd

framenet_data = framenet.exemplars()
```

```python
# エクゼンプラーを取得
corpIDs = []
count = 0

for exemplar in framenet_data:
    if "corpID" in exemplar:
        corpIDs.append(exemplar["corpID"])
    else:
        # print(framenet_data[index]["text"])
        count += 1
print(count)


exemplars = pd.DataFrame(corpIDs)
print(exemplars.value_counts())
```
