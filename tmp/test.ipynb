```python
import pandas as pd

# 辞書からDataFrameを作成
data = {
    '名前': ['太郎', '花子', '次郎', "a" ,"b" ,"c"],
    '年齢': [28, 34, 29,28,28,34],
    '職業': ['エンジニア', 'デザイナー', '医師', 'エンジニア', 'デザイナー', '医師'],
    'テスト': [["あ","a"],["い","i"],["う","u"],["あ","a"],["い","i"],["う","u"]]
}

df = pd.DataFrame(data)
print(df)
# df["test"] =  df.groupby("名前").agg(set)["職業"].apply(lambda x: len(x))
print(df.groupby("年齢")["職業"].agg(set))
print(df.groupby("年齢")["職業"].agg(set).apply(lambda x: len(x)))
print(df.groupby("年齢").agg(set)["職業"])
print(df.groupby("年齢").agg(set)["職業"].apply(lambda x: len(x)))
print()
```
